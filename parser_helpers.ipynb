```python
class Stack:
    """    Stack is implemented with a list as the storage """
    
    def __init__(self):
        """constructor"""
        self._data=[]
        
    def __len__(self):
        """returns the length of the stack"""
        return len(self._data)
    
    def is_empty(self):
        """Return True if the length is zero"""
        if len(self._data)==0:
            return True
        else:
            return False
        
    def push(self,e):
        """adds element e to the stack"""
        self._data.append(e)
        
    def top(self):
        """returns the top element"""
        return self._data[-1]

    def pop(self):
        """removes the top element"""
        return self._data.pop()
    
    def print_s(self):
        """print the stack"""
        return self._data
    
    def print_ts(self):
        """print a stack of trees"""
        for i in self._data:
            printTree(i,1,3000,18)
        return "printing"
```

```python
def init_parse():
    """initializer function that makes the stack, parse table, and production list"""
    global S, PT, P, num_errors, TS, op, stmts
    
    S = Stack()                                     # make a stack
    S.push("$")                                     # push the end of file symbol
    S.push("code")                                  # push code as the beginning non-terminal
    
    TS = Stack()                                    # tree stack
    
    op = Stack()                                    # operator stack
    
    if v: print("stack:", S.print_s(),"\n")         # print the stack
        
    PT = make_parse_table()                         # make the parse table
    P = make_productions_list()                     # make the productions list
    
    num_errors = 0                                  # variable to keep track of if there were any errors
    
    stmts = Stack()          # a stack for keeping track of the type of statement that will have curly brackets 
    stmts.push(None)
    
```

```python
def match(lookahead, non_term, m):
    """function for when non_term = lookahead
    moves on to the next token """
    global S, PT, P, t, t2
    
    if v: print("match lookahead = ", lookahead, "and top = ", non_term)
    if v: print("match stack:", S.print_s(),"\n") 
    t = t2                                             # current token is the old peak token
    t2 = lexan(m)                                      # move to the next token 
    if v: print("tokenvalue = ", t.tokenval)
```

```python
def bad_symbol(lookahead, non_term, m):
    """handle the case of no entry found in the parse table
    moves on to next token and runs the same while loop as in parser() without giving errors until we find a match"""
    global S, PT, P, t
    
    terminals = ['[', ']','++','--','=','+=','-=','*=','/=','+','-','*','/','%','(',')','<','>','<=','>=',
                 '!=','==','{','}',';','$']     # list of terminal symbols        
    if non_term in terminals:                   # check if the "bad symbol" was a terminal
        bad_production(lookahead, non_term, m)     # if it was, run bad_production where we do not skip to next token value
        return
    
    print("\nKeyError: detected symbol '", lookahead, "' on line", t.line_num, "but expected one of the following: ",non_term, end="")
    t = lexan(m)                                 # move to the next token 
    
    if t.tokentype == "OP" or t.tokentype == "RELOP" or t.tokentype == "PUN" or t.tokenval == "$":
        lookahead = t.tokenval                  
    else:
        lookahead = t.tokentype                  # lookahead is either token type or value depending on what is in the PT
        
    if v: print("\nlookahead = ", lookahead)
    
    while True:
        if v: print("bad terms stack:", S.print_s(),"\n")
        try:
            non_term = S.pop()                     # get the next non-terminal from the stack if the stack is not empty
        except IndexError:
            return                                 # if the stack is empty, handle it in parser()
        if v: print("non-term = ", non_term)
        
        if non_term == lookahead:                   # check if the lookahead matches the top value on the stack
            if non_term == "$":                     # if that value also matches the end symbol, we are done
                S.push("$")                         # put the ending symbol back on the stack to avoid index error
                return                              # go back to parser() code 
            else:                                   # otherwise, we have a match
                match(lookahead, non_term, m)
                return

        else:                                       # need to match production rule
            try:                         
                production = PT[non_term][lookahead]    # get production number from parse table
                if v: print("bad terms production number", production)
            except KeyError:
                print(",", non_term, end="")        # keep adding potential non_terminals that we are looking for to the printed error message 
                continue                            # don't try to apply production rule

            if production != 0:
                rule = P[production]                # get production rule from production list
                if rule != None:                    # check if the production sends a non-term to empty 
                    for i in range(0,len(rule)):    # otherwise, push the production onto stack in opposite order
                        S.push(rule[len(rule)-1-i])
            else:
                print(",", non_term, end="")        # keep adding to the printed error
```
